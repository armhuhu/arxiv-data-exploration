```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_pickle('dblp.pkl')
```

```python
df.shape
```

```python
df_venue = df[['id', 'venue']].groupby('venue').count()
```

```python
df_venue = df_venue.reset_index()
```

```python
ax = df.groupby('year').agg(['mean', 'median']).plot(kind='bar', figsize=(10,5))
plt.locator_params(axis='x', nbins=10)
ax.set_ylabel('Number of Citations')
ax.set_title('Average Citation by Year')
```

```python
ax = df[['year', 'id']].groupby('year').count().plot(kind='bar', figsize=(10,5))
ax.xaxis.set_major_locator(plt.MaxNLocator(10))
ax.set_title('')
```

```python
df_temp = df[['year', 'id']].groupby('year', as_index=False).count()
fig,ax = plt.subplots()
fig.set_figwidth(10)
fig.set_figheight(6)
plt.bar(df_temp.year, df_temp.id)
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
ax.set_title('Number of Papers by Year')
ax.set_ylabel('Count')
ax.set_xlabel('Year')
```
